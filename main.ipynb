```python
import pandas as pd
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BostonHousing.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop(columns='price')
features = X.columns
X = X.values
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Initilize a linear regression model and fit it to the data
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
# Visualize the how the model has assigned importance to each feature
plt.bar(features, linreg.coef_)
plt.show()
```

```python
# Evaluate the model's performance on the training set
acc = linreg.score(X_train, y_train)
acc
```

```python
import numpy as np
linreg.predict(np.array([[0.00632,	18.0,	2.31,	0,	0.538,	6.575,	65.2,	4.0900,	1,	296,	15.3,	396.90,	4.98]]))
```

```python
test_acc = linreg.score(X_test, y_test)
test_acc
```

```python
pickle.dump(linreg, open('model.pkl','wb'))
```
